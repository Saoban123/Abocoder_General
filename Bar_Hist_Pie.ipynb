### Bar Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

company = ['GOOGLE', 'AMAZIN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
plt.bar(company, revenue)
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company, revenue)
```

```python
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Technology Stocks')

plt.bar(xpos, revenue, label = 'revenue')
plt.xticks(xpos, company) # This is to specify how you want the ticks to be on the x axis e.g plt.xtixks([0.5, 10, 15])
# You use yticks to do some for the y axis

#plt.xticks([0, 5 , 10])
#xticks labels({'x = 0', 'x = 5', 'x = 10'}) use this when your tick titles have both integers and text
```

```python
# Mu
profit = [40, 2, 24, 12]
# -0.2
plt.bar(xpos, revenue, width = 0.4, label = 'Revenue', color = 'green')
plt.bar(xpos, profit, width = 0.4, label = 'Profit', color = 'purple')
# +0.2
plt.xticks(xpos, company)
plt.legend()
```

```python
profit = [40, 2, 24, 12]
# -0.2
plt.bar(xpos + 0.2, revenue, width = 0.4, label = 'Revenue', color = 'green')
plt.bar(xpos - 0.2, profit, width = 0.4, label = 'Profit', color = 'purple')
# +0.2
plt.xticks(xpos, company)
plt.legend()
```

```python
# Stacked
profit = [40, 2, 24, 12]
# -0.2
plt.barh(xpos, revenue, height = 0.4, label = 'Revenue', color = 'green')
plt.barh(xpos, profit, height = 0.4, label = 'Profit', color = 'purple')
# +0.2
plt.yticks(xpos, company)
plt.legend()
```

```python
# Horizontal multiple barchart

profit = [40, 2, 24, 12]
# -0.2
plt.barh(xpos + 0.2, revenue,  height = 0.4, label = 'Revenue', color = 'green')
plt.barh(xpos - 0.2, profit, height = 0.4, label = 'Profit', color = 'purple')
# +0.2
plt.yticks(xpos, company)
plt.legend()
```

```python
company = ['GOOGLE', 'AMAZIN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
interest = [10, 40, 30, 80]
plt.bar(company, revenue)
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company, revenue)
plt.show()
```

```python
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title('US Technology Stocks')

plt.bar(xpos, revenue, label = 'revenue')
plt.xticks(xpos, company) # This is to specify how you want the ticks to be on the x axis e.g plt.xtixks([0.5, 10, 15])
# You use yticks to do some for the y axis

plt.show()
```

```python
# Mu
profit = [40, 2, 24, 12]
# -0.2
plt.bar(xpos, revenue, width = 0.4, label = 'Revenue', color = 'green')
plt.bar(xpos, profit, width = 0.4, label = 'Profit', color = 'purple')
plt.bar(xpos, interest, width = 0.4, label = 'Interest', color = 'red')
# +0.2
plt.xticks(xpos, company)
plt.legend( bbox_to_anchor = (0, 0.5))
plt.show()
```

```python
# Mu
profit = [40, 2, 24, 12]
# -0.2
plt.bar(xpos + 0.2, revenue, width = 0.3, label = 'Revenue', color = 'green')
plt.bar(xpos, profit, width = 0.3, label = 'Profit', color = 'purple')
plt.bar(xpos - 0.2, interest, width = 0.3, label = 'Interest', color = 'red')
# +0.2
plt.xticks(xpos, company)
plt.legend()
plt.show()
```

```python
# Mu
profit = [40, 2, 24, 12]
# -0.2
plt.barh(xpos + 0.2, revenue, height = 0.3, label = 'Revenue', color = 'green')
plt.barh(xpos, profit, height = 0.3, label = 'Profit', color = 'purple')
plt.barh(xpos - 0.2, interest, height = 0.3, label = 'Interest', color = 'red')
# +0.2
plt.yticks(xpos, company)
plt.legend()
plt.show()
```

## Pie Chart

```python
exp_vals = [1400,600, 300, 410, 250]
exp_labels = [ 'Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']

plt.pie(exp_vals, labels = exp_labels)
plt.show()
```

```python
# For Perfect circle

plt.pie(exp_vals, labels = exp_labels, shadow = True)
plt.axis('equal')
plt.show()
```

```python
plt.pie(exp_vals, labels = exp_labels, shadow = True, autopct = '%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
plt.pie(exp_vals, labels = exp_labels, shadow = True, autopct = '%1.1f%%', explode = [0, 0, 0, 0.1, 0.2])
plt.axis('equal')
plt.show()
```

```python
Mycolor = ['green', 'cyan', 'coral', 'purple', 'hotpink', 'wine']
plt.pie(exp_vals, labels = exp_labels, shadow = True, autopct = '%1.1f%%',radius = 2.0, explode = [0, 0, 0, 0.1, 0.2], colors = Mycolor, startangle = 270)
plt.axis('equal')

plt.legend(title = 'Abocoder', bbox_to_anchor = (1, 1))
plt.show()

```
