```python
import pandas as pd
indian_weather = pd.DataFrame({
    'city' : ['Mumbai', 'Delhi', 'Banglore'],
    'humidity' : [80, 60, 78],
    'temperature' : [32, 45, 30]
})
indian_weather
```

```python
import pandas as pd
Students = pd.DataFrame({
    'Name' : ['saoban', 'lateefat', 'simi', 'pecu', 'Joy'],
    'Sex' : ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Age' : [20, 14, 17, 19, 12],
    'Occupation' :['Data Scientist', 'Doctor','Lawyer', 'Engineer', 'Geologist']
})

Students
```

```python
# Another data set
add = pd.DataFrame({
    'Address' : ['isolu', 'camp', 'oluwo', 'zoo', 'gate'],
    'Phone number' : [1234, 5678, 9876, 2341, 3898]
})
add
```

```python
# Concatenation of the two sets
# axis = 0 is for row
# axis = 1 is for column
df = pd.concat([Students, add], axis = 1)
df
```

```python
Nigeria_weather = pd.DataFrame({
    'city' : ['Lagos', 'Ogun', 'Osun'],
    'humidity' : [70, 40, 90],
    'temperature' : [23, 35, 73]
})
Nigeria_weather
```

```python
df = pd.concat([indian_weather, Nigeria_weather])
df
```

```python
# To ignore the index
ddf = pd.concat([indian_weather, Nigeria_weather], ignore_index = True)
ddf
```

```python
# Specifying the keys
df = pd.concat([indian_weather, Nigeria_weather],  keys = ['Indian', 'Nigeria'])
df
```

```python
df.loc['Nigeria']
```

```python
# Creating a series
s = pd.Series(['Humidity', 'Dry', 'Rain'], name = 'event')
s
```

```python
con = pd.concat([ddf, s], axis = 1)
con
```
