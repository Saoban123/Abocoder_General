```python
# Using the replace function

Replace = pd.DataFrame({
    'Day' : ['1/1/17', '2/1/17', '3/1/17', '4/1/17', '5/1/17'],
    'Temperature' : [32, -9999, '28c', -9999, -8888],
    'WindSpeed' : ['6mph', '7mph', -9999, 7, 5],
    'Events' : ['Rain', 'Sunny', 'Snow', 'No Event', 'Rain']
})

Replace
```

```python
rr = Replace.replace(-9999, value = np.NaN)
rr
```

```python
# To replace a list
lst = Replace.replace(to_replace = [-9999, -8888], value = 0)
lst
```

```python
# To replace with different value

Plenty_value = Replace.replace({
    -9999 : np.NaN,
    -8888 : 0,
    'No Event' : 'Sunny'
    
})

Plenty_value
```
