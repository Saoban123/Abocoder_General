```python
# N:B -- In python the date writing date starts with year/month/day
import pandas as pd

weather = pd.read_csv('weather_data_new.csv', parse_dates = ['day'])

weather
```

```python
# Forward fill
# This fill the missing rows with the values above it

ffill_weather_data = weather.fillna(method = 'ffill')
ffill_weather_data
```

```python
# Backwark fill
# This fill wwith the value below

bfill_weather_data = weather.fillna(method = 'bfill')
bfill_weather_data
```

```python
# Specifying axis
# This fill it in with the one beside it

Sffill_weather_data = weather.fillna(method = 'ffill', axis = 1)
Sffill_weather_data
```

```python
Sbfill_weather_data = weather.fillna(method = 'bfill', axis = 1)
Sbfill_weather_data
```

```python
# Limit Parameter
# This just specify the number of NaN it has to fill
Lbfill_weather_data = weather.fillna(method = 'bfill', limit = 1)
Lbfill_weather_data
```

```python
# Specifying axis
# This fill it in with the one beside it

Lffill_weather_data = weather.fillna(method = 'ffill', limit = 1)
Lffill_weather_data
```

```python
# Filling in with middle ones

Interpolate = weather.interpolate()
Interpolate
```

```python
# Calculating the in between number using the day as a guide

Index_weather = weather.set_index('day')
Index_weather
```

```python
timeInterpolate = Index_weather.interpolate(method = 'time')
timeInterpolate
```

```python
# Dropna function
# Drops all rows and column where there is NaN

Drop_weather = weather.dropna()
Drop_weather
```

```python
# How function is used to drop the one one wants

How = Index_weather.dropna(how = 'all')
How
```

```python
# This is just like the normal dropna 

Any = weather.dropna(how = 'any')
Any
```

```python
Thresh = Index_weather.dropna(thresh = 1)
Thresh
```
